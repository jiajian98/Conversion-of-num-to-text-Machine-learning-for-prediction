```python
%pip install seaborn
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('./bicycle/bike_sharing.csv')
df.head()
df.info()
```

```python
#check if data cleaning required
df.isnull().sum()
df.duplicated().sum()
```

```python
#This histogram is to check how cnts are distributed
plt.figure(figsize=(8,5))
sns.histplot(df['cnt'],bins=40,kde=True)
plt.title('Distribution of total rental bikes (cnt)')
plt.xlabel("cnt")
plt.ylabel("frequency")
plt.axvline(df['cnt'].mean(),color='green',linestyle='--',label='Mean')
plt.legend()
plt.show()
```

```python
#This lineplot to check the trends for rental bikes per hour
plt.figure(figsize=(7,5))
sns.lineplot(x='hr',y='cnt',data=df)
plt.title('Average Rental Bikes per hour')
plt.xlabel("Hour")
plt.ylabel("Average Rentals")
plt.show()
```

```python
#This bar chart is to check the average rental bikes across month
plt.figure(figsize=(8,5))
sns.barplot(x='mnth',y='cnt',data=df)
plt.title('Average Rental Bikes per month')
plt.xlabel("Month")
plt.ylabel("Average Rentals")
plt.show()

```

```python
#This bar chart is to compare the average rental bikes between years
plt.figure(figsize=(8,5))
sns.barplot(x='yr',y='cnt',data=df)
plt.title('Average Rental Bikes per year')
plt.xlabel("Year")
plt.ylabel("Average Rentals")
plt.show()
```

```python
#This bar chart is to check the average rental bikes on weekday
plt.figure(figsize=(8,5))
sns.barplot(x='weekday',y='cnt',data=df)
plt.title('Average Rental Bikes per weekday (0=Sunday)')
plt.xlabel("Weekday")
plt.ylabel("Average Rentals")
plt.show()
```

```python
#This bar chart is to compare the rental bikes on working day and non-working day
plt.figure(figsize=(8,5))
sns.barplot(x='workingday',y='cnt',data=df,)
plt.title('Average Rental Bikes on working day(1) and non-working day(0)')
plt.xlabel("Working Day")
plt.ylabel("Average Rentals")
plt.show()
```

```python
#This bar chart is to compare the rental bikes on holidays and non-holidays
plt.figure(figsize=(8,5))
sns.barplot(x='holiday',y='cnt',data=df,)
plt.title('Average Rental Bikes on holiday(1) and non-holiday(0)')
plt.xlabel("Holiday")
plt.ylabel("Average Rentals")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='weathersit',y='cnt',data=df)
plt.title('Average Rental Bikes by Weather Situation')
plt.xlabel('Weather Situation')
plt.ylabel('Average Rentals')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='season',y='cnt',data=df)
plt.title('Average Rental Bikes by Season')
plt.xlabel('Season')
plt.ylabel('Average Rentals')
plt.show()
```

```python
#These scatterplots explore how key continuous feature affect the rental bikes
#Temperature vs Bike rentals
plt.figure(figsize=(8,5))
sns.scatterplot(x='temp',y='cnt',data=df)
plt.title('Average Rental Bikes by Temperature')
plt.xlabel('Temperature')
plt.ylabel('Average Rentals')
plt.show()
```

```python
#Feeling Temperature vs Bike rentals
plt.figure(figsize=(8,5))
sns.scatterplot(x='atemp',y='cnt',data=df)
plt.title('Average Rental Bikes by Feeling Temperature')
plt.xlabel('Feeling Temperature')
plt.ylabel('Average Rentals')
plt.show()
```

```python
#Humidity vs Bike rentals
plt.figure(figsize=(8,5))
sns.scatterplot(x='hum',y='cnt',data=df)
plt.title('Average Rental Bikes by Humidity')
plt.xlabel('Humidity')
plt.ylabel('Average Rentals')
plt.show()
```

```python
#Windspeed vs Bike rentals
plt.figure(figsize=(8,5))
sns.scatterplot(x='windspeed',y='cnt',data=df,alpha=0.5)
plt.title('Average Rental Bikes by Windspeed')
plt.xlabel('Windspeed')
plt.ylabel('Average Rentals')
plt.show()
```

```python
#To visualize correlation between all variables
plt.figure(figsize=(8,5))
corr = df.corr()
sns.heatmap(corr, annot=True, fmt='.2f')
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
#EDA Summary
#1.Bike rentals peak during commute hour, specically around 8am and 5-6pm, with overall demand rising from May and gradually declining after October.
#2.Bike rental in 2012 were higher than in 2011.
#3.Weekday show higher demand on workings day compared to weekends/holidays, reflecting commuter behaviour.
#4.Clear weather and moderate temperature increase bike usage.
#5.Summer and fall seasons show higher bike usage as compared to winter and spring.
#6.The temp and atemp are highly correlated. Therefore only temp will be used for the modelling.
#7.High humidity and high windspeed reduces bike rentals.
#8.It can conclude that the hours,weather,season and temperature are the most influential factors affecting bike rental demand.
```
